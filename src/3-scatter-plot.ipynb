### Exercise 3: Scatter Plot with Annotations

```python
# TASK: Create a scatter plot with x = [1, 2, 3, 4, 5] and y = [2, 4, 6, 8, 10].
# Annotate each point with its (x, y) value, and set the title as 'Scatter Plot with Annotations'.
```

```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Create the scatter plot
plt.scatter(x, y, color='blue')

# Annotate each point with its (x, y) value
# (Using a small offset so the text doesn't overlap the marker)
for x_val, y_val in zip(x, y):
    plt.annotate(
        f"({x_val}, {y_val})", 
        (x_val, y_val),
        textcoords="offset points", 
        xytext=(0, 5),   # Move the text 5 points above the marker
        ha='center'
    )

# Add a title
plt.title("Scatter Plot with Annotations")

# (Optional) Label the axes
plt.xlabel("X")
plt.ylabel("Y")

# Show the plot
plt.show()
```
